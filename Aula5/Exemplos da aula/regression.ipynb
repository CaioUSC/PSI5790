```python
#regression.py - 2024
import os; os.environ['TF_CPP_MIN_LOG_LEVEL']='3'
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Input
from tensorflow.keras import optimizers
import numpy as np

#Define modelo de rede
model = Sequential(); model.add(Input(shape=(2,)))
#camada densa todas entradas estão conectadas com todos neuronios
model.add(Dense(2, activation='sigmoid'))
#camada sem ativação para fazer a regressão
model.add(Dense(2, activation=None))

# Otimizador e quem vai mexer nos pesos e no BIAS para chegar no resultado
# learning_rate e quanto iremos alterar os parametros
sgd=optimizers.SGD(learning_rate=1.0)
# loss -> como iremos calcular o erro
# mean square error
model.compile(optimizer=sgd,loss='mse')

AX = np.matrix('0.9 0.1; 0.1 0.9',dtype='float32')
AY = np.matrix('0.1 0.9; 0.9 0.1',dtype='float32')
print("AX"); print(AX)
print("AY"); print(AY)

# treino da rede neural usando 100 epocas ira repetir 100 vezes
# As opcoes sao usar batch_size=2 ou 1 , processar em lotes de quantos
# sendo batch_size = 2 ele encontra a media dos erros de 2 entradas e tenta ajustar
# verbose 0 -> nada 1-> tudo 2-> so partes (descrição do processo)
model.fit(AX, AY, epochs=100, batch_size=2, verbose=2)

QX = np.matrix('0.9 0.1; 0.1 0.9; 0.8 0.0; 0.2 0.9',dtype='float32')
print(QX)
QP=model.predict(QX, verbose=2)
print(QP)
```
